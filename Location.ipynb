## H3 plots with Folium

```python
from h3_map import h3_map
```

### Pandas process

```python
import pandas as pd
```

```python
h3 = h3_map('pandas')
pd_df = pd.read_csv("french_locations.csv")
```

```python
pd_hex = h3.to_hex_id(pd_df, 'lat', 'long', resolution=4)
```

```python
pd_agg = pd_hex.groupby('hex_id').count().reset_index()
pd_agg.rename(columns={'lat':'value'}, inplace=True)
pd_agg = pd_agg[['hex_id','value']]
```

```python
h3.h3_folium_map(pd_agg)
```

### Spark process

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('H3_Folium').getOrCreate()
spark
```

```python
h3 = h3_map('spark')
s_df = spark.read.csv("french_locations.csv", header=True)
```

```python
s_hex = h3.to_hex_id(s_df, 'lat', 'long', resolution=4)
```

```python
s_agg = s_hex.groupBy('hex_id').count().withColumnRenamed('count','value')
```

```python
h3.h3_folium_map(s_agg)
```
