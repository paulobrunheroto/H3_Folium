## H3 plots with Folium

```python
from h3_folium.h3_python import PythonH3
from h3_folium.h3_spark import SparkH3
```

### Pandas process

```python
import pandas as pd
```

```python
pd_df = pd.read_csv("input_files/french_locations.csv")
```

```python
ph3 = PythonH3(
    lat_col='lat',
    long_col='long',
    resolution=4
)
```

```python
ph3.map_gen(pd_df, 'pandas_folium')
```
